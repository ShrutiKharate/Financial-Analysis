```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
```

```python
df = pd.read_csv("Finance_data.csv")
df
```

```python
# 1. Age Distribution Count Plot

sns.countplot(x='age', data=df, palette="Set3", linewidth=2, edgecolor='black', hue='age', dodge=False, legend=False)
plt.show()
```

```python
# 2. Objective vs Expected Returns

plt.figure(figsize=(12, 6))
sns.boxplot(x='Objective', y='Expect', data=df)
plt.title('Investment Objective vs Expected Returns')
plt.xlabel('Investment Objective')
plt.ylabel('Expected Returns')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3. Purpose of Investment by Gender

plt.figure(figsize=(12, 6))
purpose_gender = pd.crosstab(df['Purpose'], df['gender'], normalize='index')
purpose_gender.plot(kind='bar', stacked=True)
plt.title('Purpose of Investment by Gender')
plt.xlabel('Purpose')
plt.ylabel('Proportion')
plt.legend(title='Gender')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 4. Factor Influencing Investment Decision by Age

g = sns.FacetGrid(df, col='Factor', col_wrap=3, height=4)
g.map(sns.histplot, 'age')
g.add_legend()
plt.suptitle('Age Distribution by Factor Influencing Investment Decision', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# 5. Word Cloud of Investment Avenues

from wordcloud import WordCloud

plt.figure(figsize=(12, 8))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['Avenue']))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Preferred Investment Avenues')
plt.tight_layout()
plt.show()
```

```python
df.drop(['Mutual_Funds', 'Equity_Market',
       'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold'],axis=1,inplace=True)
```

```python
# 6. Gender Distribution Count Plot

sns.countplot(df['gender'],linewidth=3,palette="Set2",edgecolor='black')
plt.show()

```

```python
# 7. Age Distribution Count Plot (Repeated)

sns.countplot(x='age', data=df, palette="Set3", linewidth=2, edgecolor='black', hue='age', dodge=False, legend=False)
plt.show()
```

```python
# 8. Investment Avenues and Stock Market by Gender

plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(x=df['gender'],hue=df['Investment_Avenues'],palette='summer',linewidth=3,edgecolor='white')
plt.title('INVESTMENT AVENUE')
plt.subplot(1, 2, 2)
sns.countplot(x=df['gender'],hue=df['Stock_Marktet'],palette='hot',linewidth=3,edgecolor='white')
plt.title('STOCK MARKET')
plt.show()


# Mostly females dont invest or go to stocks 
```

```python
# 9. Factors Affecting Investing

sns.countplot(x=df['Factor'],palette='coolwarm',linewidth=2,edgecolor='black')

# Returns is the most influential Factor considered while investing which is followed by risk 
```

```python
# 10. Risk-Taking Behavior by Gender

sns.countplot(x=df['gender'],hue=df['Factor'],palette='Oranges',linewidth=2,edgecolor='black')
plt.title('Gender')
plt.show()
```

```python
# 11. Purpose of Investment vs Age

plt.figure(figsize=(16,6))
sns.pointplot(x="Purpose",y='age',data=df,linestyles="--",capsize=.3,color='Red')
plt.show()
```

```python
# 12. Duration and Investment Monitoring by Gender

plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(hue=df['gender'],x=df['Duration'],palette='viridis',linewidth=2,edgecolor='black')
plt.title('DURATION')
plt.subplot(1, 2, 2)
sns.countplot(hue=df['gender'],x=df['Invest_Monitor'],palette='seismic',linewidth=2,edgecolor='black')
plt.title('INVESTMENT MONITORING')
plt.show()

#Below graphs show us that most of the people invest from 1-5 Years and monitor their investments mostly monthly
```
